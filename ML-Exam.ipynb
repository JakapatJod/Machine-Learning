```python
import os
# เปลี่ยนชื่อรูปภาพ
def rename_images(directory):
    if not os.path.isdir(directory):
        print(f"Error: Directory '{directory}' not found.")
        return

    image_files = [f for f in os.listdir(directory) if os.path.isfile(os.path.join(directory, f))]
    num_images = len(image_files)

    # เปลี่ยนชื่อของแต่ละรูปภาพ
    for i, image_name in enumerate(image_files, start=0):
        # ดึงนามสกุลไฟล์ของรูปภาพ
        _, ext = os.path.splitext(image_name)
        # สร้างชื่อใหม่โดยใช้รูปแบบ "{เลขลำดับ}.{นามสกุลไฟล์}"
        # new_name = f"{i}{ext}" #  new_name = f"y{i}{ext}" y1.png
        new_name = f"{i}.jpg"


        # เปลี่ยนชื่อของไฟล์
        os.rename(os.path.join(directory, image_name), os.path.join(directory, new_name))
        print(f"Renamed '{image_name}' to '{new_name}'.")

# เรียกใช้ฟังก์ชันเพื่อเปลี่ยนชื่อรูปภาพใน directory ที่กำหนด
rename_images(r"C:\Users\ADMIN\Desktop\flowers\sunflower")
```
