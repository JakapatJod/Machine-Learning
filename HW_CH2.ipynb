```python
import re
from matplotlib import pyplot as plt
from nltk.tokenize import regexp_tokenize
from collections import defaultdict
```

```python
import io
from google.colab import files

# Upload the file
uploaded = files.upload()

# Read the content of the uploaded file
with io.StringIO(uploaded['holy_grail.txt'].decode('utf-8')) as f:
    holy_grail = f.read()

# Display the content
print(holy_grail)
```

```python
lines = holy_grail.split('\n')

# Create a defaultdict to store letter counts
letter_counts = defaultdict(int)

pattern = r"\w+[a-zA-Z]"

# Count words and their starting letters
for line in lines:
    for word in regexp_tokenize(line,pattern ):
        first_letter = word[0].upper()
        letter_counts[first_letter] += 1


# Extract letters and counts
letters = sorted(letter_counts.keys())
counts = [letter_counts[letter] for letter in letters]


print(lines)
for letter in sorted(letter_counts.keys()):
    print(f"{letter}: {letter_counts[letter]}")


# Create the bar chart
plt.bar(letters, counts)
plt.xlabel('Letter')
plt.ylabel('Count')
plt.title('Word Counts Starting with A-Z')
plt.show()
```
